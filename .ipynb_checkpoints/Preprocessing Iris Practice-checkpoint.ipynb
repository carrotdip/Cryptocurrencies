```python
# Import Dependencies
import pandas as pd
```

```python
# Read Iris CSV file
file_path = "Resources/iris.csv"
iris_df = pd.read_csv(file_path)
iris_df.head()
```

```python
# Drop class column as unsupervised learning only deals with numerical data
iris_df = iris_df.drop(['class'], axis=1)
iris_df.head()
```

```python
# Reorder columns so that sepal and petal lengths are the first two columns and width is last
iris_df = iris_df[['sepal_length','petal_length','sepal_width','petal_width']]
iris_df
```

```python
# Create CSV file of preprocessed data
output_file_path = "Resources/new_iris_data.csv"
iris_df.to_csv(output_file_path, index=False)
```
